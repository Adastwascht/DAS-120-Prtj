```python
fire_countDF  = pd.read_excel("1990 - 2023 Avrupa Yangın Alanı ve Miktarı.xlsx", header= 0, sheet_name=0).iloc[10:].reset_index(drop=True) # This deletion is related to the excel file I have last column was extra
fire_areaDF  = pd.read_excel("1990 - 2023 Avrupa Yangın Alanı ve Miktarı.xlsx", header= 0, sheet_name=1).iloc[10:].reset_index(drop=True) # The unit of this data is hectares
total_yearly_forest_area = pd.read_csv("Total Forest Area.csv").iloc[:-6,2:] #The unit of this data is kmsq
```

```python
total_yearly_forest_area = total_yearly_forest_area.drop(columns=["Time Code"])
total_yearly_forest_area = total_yearly_forest_area.rename(columns={"Time":"year","Algeria [DZA]":"Algeria", "Cyprus [CYP]":"Cyprus", "France [FRA]":"France", "Greece [GRC]":"Greece","Italy [ITA]":"Italy" , "Lebanon [LBN]":"Lebanon", "Portugal [PRT]":"Portugal", "Spain [ESP]":"Spain", "Turkiye [TUR]":"Turkey"})

#One kmsq is 100 hectares (we also need to change object type to numeric type)
total_yearly_forest_area = total_yearly_forest_area.apply(pd.to_numeric)
```

```python
total_yearly_forest_area.iloc[:,1:] = total_yearly_forest_area.iloc[:,1:] *100
total_yearly_forest_area.info()
total_yearly_forest_area
```

```python
fire_areaDF.info()
fire_countDF.info()
```

We need to convert the object types to numeric. After that, we are interested total burnt area / total forest area, in order to make things comparable

```python
fire_areaDF = fire_areaDF.apply(pd.to_numeric, errors = "coerce") # "coerce" to ignore NaN values
fire_countDF = fire_countDF.apply(pd.to_numeric, errors = "coerce")
fire_areaDF = fire_areaDF.rename(columns={"DZA":"Algeria", "CYP":"Cyprus", "FRA":"France", "GRC":"Greece","ITA":"Italy" , "LBN":"Lebanon", "PRT":"Portugal", "ESP":"Spain", "TUR":"Turkey"})
fire_countDF = fire_countDF.rename(columns={"DZA":"Algeria", "CYP":"Cyprus", "FRA":"France", "GRC":"Greece","ITA":"Italy" , "LBN":"Lebanon", "PRT":"Portugal", "ESP":"Spain", "TUR":"Turkey"})
```

```python
fire_areaDF.info()
fire_areaDF
```

```python
fire_countDF.info()
fire_countDF
```

```python
fire_area_ratioDF = fire_areaDF.copy()
fire_area_ratioDF.iloc[:,1:] = fire_areaDF.iloc[:,1:].div(total_yearly_forest_area.iloc[:,1:].replace(0,pd.NA))

fire_count_ratioDF = fire_countDF.copy()
fire_count_ratioDF.iloc[:,1:] = fire_countDF.iloc[:,1:].div(total_yearly_forest_area.iloc[:,1:].replace(0,pd.NA))
```

```python
fire_area_ratioDF.iloc[:,1:] = fire_area_ratioDF.iloc[:,1:] * 100 # Converting to percentage for readibility
fire_count_ratioDF.iloc[:,1:] = fire_count_ratioDF.iloc[:,1:] * 100
fire_area_ratioDF
```

```python
fire_count_ratioDF
```

```python
fire_ratio_elaborateDF = fire_area_ratioDF.melt(id_vars="year", var_name="country", value_name="burned_ratio")
fire_count_elaborateDF = fire_countDF.melt(id_vars="year", var_name="country", value_name="fire_count")
fire_count_ratio_elaborateDF = fire_count_ratioDF.melt(id_vars="year", var_name="country", value_name="fire_incidence")


fire_ratio_elaborateDF
```

```python
fire_ratio_elaborateDF.to_csv("Final Form Elaborate Fire Ratio.csv", index=False)
fire_count_elaborateDF.to_csv("Final Form Elaborate Fire Count.csv", index=False)
fire_count_ratio_elaborateDF.to_csv("Final Form Elaborate Fire Count Ratio.csv", index=False)
```
